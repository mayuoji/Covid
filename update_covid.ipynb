```python
#load data
import csv
import requests
import io
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

#import data
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
r = requests.get(url)
df = pd.read_csv(io.BytesIO(r.content))
df=df[['location','date','new_cases','new_deaths','new_cases_smoothed','new_deaths_smoothed','new_cases_per_million','new_deaths_per_million','new_cases_smoothed_per_million','new_deaths_smoothed_per_million']]
df.tail()
```

## Major_newcase

```python
##major
df1=df[(df['location']=="Spain")|(df['location']=="Italy")|(df['location']=="France")|(df['location']=="United Kingdom")|(df['location']=="Germany")]

##date range
df1['date']=pd.to_datetime(df1['date'],format='%Y-%m-%d')
df1=df1[df1['date'] >= dt.datetime(2020,2,1)] 

#plot
df1.groupby(['date','location']).sum()['new_cases_smoothed'].unstack().plot()

#for save file format
fig=df1.groupby(['date','location']).sum()['new_cases_smoothed'].unstack().plot()
plt.savefig("img1.png")

#just for hide warinig error (sorry..)
import warnings
warnings.filterwarnings('ignore')
```
